```python
import pandas as pd
```

```python
df = pd.read_csv('./iris.data')
df.head()
```

```python
df['Name'] = df['Name'].str[5:]
```

```python
df
```

# Group by Iris type

```python
df.groupby('Name').mean()
```

```python
df.plot?
```

```python
colors = {'setosa':'red', 'versicolor':'blue', 'virginica':'black'}
df.plot(kind='scatter', x='SepalLength', y='SepalWidth', c=df['Name'].apply(lambda x: colors[x]))
```

```python
from pandas.tools.plotting import radviz
plt.figure(figsize=(8,6))
radviz(df, 'Name')
```
